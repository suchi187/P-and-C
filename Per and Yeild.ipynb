```python
df=df[['inning','over','batsman','bowler','batsman_runs','extra_runs','total_runs']]
df
```

```python
cat_features = ['batsman','bowler']
df = pd.get_dummies(df,columns=cat_features)
print(df.columns)
```

```python
from sklearn.preprocessing import StandardScaler
# finding the mean and std deviation of numerical columns
scaler=StandardScaler().fit(df[['inning','over','batsman_runs','extra_runs','total_runs']]) 
# scaling columns to a common range
df_normalized = scaler.transform(df[['inning','over','batsman_runs','extra_runs','total_runs']])# Scaled data 
#note that output is a 2-Dimensional array of normalized data
df_normalized
```

```python
#updating the numerical(normalized) data into the dataframe
df_normalized=pd.DataFrame(df_normalized,columns=['inning','over','batsman_runs','extra_runs','total_runs'])
df_normalized=df_normalized.join(df[df.columns.drop(['inning','over','batsman_runs','extra_runs','total_runs'])])
```

```python
Y=df_normalized['over']
X=df_normalized[df_normalized.columns.drop('over')]
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=100)
X_train.shape , X_test.shape
```

```python
#importing required module to build the model
from sklearn.linear_model import LinearRegression
#building and training the model
model = LinearRegression()
model.fit(X_train,Y_train)
#Evaluating the model on the train and test data for R-Squared score 
train_score=model.score(X_train,Y_train)
test_score=model.score(X_test,Y_test)
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```

```python
from sklearn.metrics import mean_squared_error
train_predictions = model.predict(X_train)
train_RMSE=mean_squared_error(Y_train,train_predictions)**0.5
test_predictions = model.predict(X_test)
test_RMSE=mean_squared_error(Y_test,test_predictions)**0.5
print('Train RMSE : ',train_RMSE)
print('Test RMSE : ',test_RMSE)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data=pd.read_csv("D:/aiml/yield_df.csv")
data 
```

```python
print(data.shape)
```

```python
print(data.info())
```

```python
data=data[['Area','Item','Year','hg/ha_yield','average_rain_fall_mm_per_year','pesticides_tonnes','avg_temp']]
data
```

```python
# encoding categorical values
cat_features=['Area','Item']
data= pd.get_dummies(data,columns=cat_features)
print(data.columns)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(data[['Year','hg/ha_yield','average_rain_fall_mm_per_year','pesticides_tonnes','avg_temp']]) 
data_normalized = scaler.transform(data[['Year','hg/ha_yield','average_rain_fall_mm_per_year','pesticides_tonnes','avg_temp']]) 
data_normalized
```

```python
#updating the numerical(normalized) data into the dataframe
data_normalized=pd.DataFrame(data_normalized,columns=['Year','hg/ha_yield','average_rain_fall_mm_per_year','pesticides_tonnes','avg_temp'])
data_normalized=data_normalized.join(data[data.columns.drop(['Year','hg/ha_yield','average_rain_fall_mm_per_year','pesticides_tonnes','avg_temp'])])
```

```python
Y=data_normalized['Year']
X=data_normalized[data_normalized.columns.drop('Year')]
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=100)
X_train.shape , X_test.shape

```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
train_score=model.score(X_train,Y_train)
test_score=model.score(X_test,Y_test)
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```

```python
from sklearn.metrics import mean_squared_error
train_predictions = model.predict(X_train)
train_RMSE=mean_squared_error(Y_train,train_predictions)**0.5
test_predictions = model.predict(X_test)
test_RMSE=mean_squared_error(Y_test,test_predictions)**0.5
print('Train RMSE : ',train_RMSE)
print('Test RMSE : ',test_RMSE)
```
